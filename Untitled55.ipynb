```python
import torch
import numpy as np

import pytorch_lightning as pl

import torch.nn.functional as F

from torchmetrics.classification import (
    MulticlassAccuracy,
    MulticlassPrecision,
    MulticlassRecall,
    MulticlassF1Score,
    MulticlassAUROC
)

from torch_geometric.loader import DataLoader
from torch_geometric.nn import (
    GCNConv,
    global_mean_pool
)
```

```python
TRAIN_PATH = "/content/drive/MyDrive/endocellgraph/cache/train_graphs.pt"
VAL_PATH = "/content/drive/MyDrive/endocellgraph/cache/val_graphs.pt"
```

```python
train_graphs = torch.load(
    TRAIN_PATH,
    weights_only=False
)
```

```python
val_graphs = torch.load(
    VAL_PATH,
    weights_only=False
)
```

```python
labels = []

for graph in train_graphs:
    labels.append(int(graph.y))

num_classes = len(set(labels))
```

```python
num_classes
```

```python
train_loader = DataLoader(
    train_graphs,
    batch_size=32,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)
```

```python
val_loader = DataLoader(
    val_graphs,
    batch_size=32,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)
```

```python
import torch

class PatchGCN(torch.nn.Module):

    def __init__(self, num_classes):

        super().__init__()

        self.conv1 = GCNConv(
            768,
            256
        )

        self.conv2 = GCNConv(
            256,
            128
        )

        self.conv3 = GCNConv(
            128,
            64
        )

        self.fc = torch.nn.Linear(
            64,
            num_classes
        )

    def forward(
        self,
        x,
        edge_index,
        batch
    ):

        x = F.relu(
            self.conv1(
                x,
                edge_index
            )
        )

        x = F.dropout(
            x,
            p=0.3,
            training=self.training
        )

        x = F.relu(
            self.conv2(
                x,
                edge_index
            )
        )

        x = F.dropout(
            x,
            p=0.3,
            training=self.training
        )

        x = F.relu(
            self.conv3(
                x,
                edge_index
            )
        )

        x = global_mean_pool(
            x,
            batch
        )

        return self.fc(x)
```

```python
class LitGCN(pl.LightningModule):

    def __init__(
        self,
        num_classes
    ):

        super().__init__()

        self.model = PatchGCN(
            num_classes
        )

        self.acc = MulticlassAccuracy(
            num_classes=num_classes
        )

        self.precision = MulticlassPrecision(
            num_classes=num_classes,
            average="macro"
        )

        self.recall = MulticlassRecall(
            num_classes=num_classes,
            average="macro"
        )

        self.f1 = MulticlassF1Score(
            num_classes=num_classes,
            average="macro"
        )

        self.auc = MulticlassAUROC(
            num_classes=num_classes,
            average="macro"
        )

    def forward(
        self,
        x,
        edge_index,
        batch
    ):

        return self.model(
            x,
            edge_index,
            batch
        )

    def training_step(
        self,
        batch,
        batch_idx
    ):

        logits = self(
            batch.x,
            batch.edge_index,
            batch.batch
        )

        loss = F.cross_entropy(
            logits,
            batch.y
        )

        self.log(
            "train_loss",
            loss,
            prog_bar=True
        )

        return loss

    def validation_step(
        self,
        batch,
        batch_idx
    ):

        logits = self(
            batch.x,
            batch.edge_index,
            batch.batch
        )

        probs = torch.softmax(
            logits,
            dim=1
        )

        preds = torch.argmax(
            probs,
            dim=1
        )

        self.acc.update(
            preds,
            batch.y
        )

        self.precision.update(
            preds,
            batch.y
        )

        self.recall.update(
            preds,
            batch.y
        )

        self.f1.update(
            preds,
            batch.y
        )

        self.auc.update(
            probs,
            batch.y
        )

    def on_validation_epoch_end(self):

        self.log(
            "val_acc",
            self.acc.compute(),
            prog_bar=True
        )

        self.log(
            "val_precision",
            self.precision.compute(),
            prog_bar=True
        )

        self.log(
            "val_recall",
            self.recall.compute(),
            prog_bar=True
        )

        self.log(
            "val_f1",
            self.f1.compute(),
            prog_bar=True
        )

        self.log(
            "val_auc",
            self.auc.compute(),
            prog_bar=True
        )

        self.acc.reset()
        self.precision.reset()
        self.recall.reset()
        self.f1.reset()
        self.auc.reset()

    def configure_optimizers(self):

        return torch.optim.AdamW(
            self.parameters(),
            lr=1e-3,
            weight_decay=1e-4
        )
```

```python
from pytorch_lightning.callbacks import (
    ModelCheckpoint
)

checkpoint_callback = ModelCheckpoint(
    monitor="val_acc",
    mode="max",
    save_top_k=1,
    filename="best-gcn"
)
```

```python
model = LitGCN(
    num_classes=num_classes
)
```

```python
trainer = pl.Trainer(
    max_epochs=50,
    accelerator="gpu",
    devices=1,
    callbacks=[checkpoint_callback],
    log_every_n_steps=10
)
```

```python
trainer.fit(
    model,
    train_loader,
    val_loader
)
```

```python
best_model = LitGCN.load_from_checkpoint(
    checkpoint_callback.best_model_path,
    num_classes=num_classes
)
```

```python
best_model.eval()
best_model.cuda()
```

```python
import torch
import numpy as np

all_probs = []
all_preds = []
all_labels = []

with torch.no_grad():

    for batch in val_loader:

        batch = batch.to("cuda")

        logits = best_model(
            batch.x,
            batch.edge_index,
            batch.batch
        )

        probs = torch.softmax(
            logits,
            dim=1
        )

        preds = probs.argmax(dim=1)

        all_probs.append(
            probs.cpu()
        )

        all_preds.append(
            preds.cpu()
        )

        all_labels.append(
            batch.y.cpu()
        )

all_probs = torch.cat(all_probs).numpy()
all_preds = torch.cat(all_preds).numpy()
all_labels = torch.cat(all_labels).numpy()
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

metrics = {
    "Accuracy":
        accuracy_score(
            all_labels,
            all_preds
        ),

    "Precision":
        precision_score(
            all_labels,
            all_preds,
            average="macro"
        ),

    "Recall":
        recall_score(
            all_labels,
            all_preds,
            average="macro"
        ),

    "F1":
        f1_score(
            all_labels,
            all_preds,
            average="macro"
        ),

    "ROC-AUC":
        roc_auc_score(
            all_labels,
            all_probs,
            multi_class="ovr",
            average="macro"
        )
}

import pandas as pd

pd.DataFrame(
    metrics,
    index=["Value"]
)
```

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        all_labels,
        all_preds
    )
)
```
