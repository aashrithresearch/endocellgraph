```python
from datasets import load_dataset

dataset = load_dataset(
    '/content/drive/MyDrive/endocellgraph'
)
```

```python
splits = dataset["train"].train_test_split(
    test_size=0.1,
    seed=42,
    stratify_by_column="label"
)
```

```python
train_ds = splits["train"]
val_ds = splits["test"]
```

```python
import torch
from transformers import AutoImageProcessor, AutoModel
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
processor = AutoImageProcessor.from_pretrained(
    "facebook/dinov2-base"
)

backbone = AutoModel.from_pretrained(
    "facebook/dinov2-base"
)

backbone = backbone.to(device)
backbone.eval()
```

```python
next(backbone.parameters()).device
```

```python
import torch

def create_grid_edges():
    edges = []

    for r in range(14):
        for c in range(14):
            node = r * 14 + c

            for dr in [-1, 0, 1]:
                for dc in [-1, 0, 1]:

                    if dr == 0 and dc == 0:
                        continue

                    nr, nc = r + dr, c + dc

                    if 0 <= nr < 14 and 0 <= nc < 14:
                        neigh = nr * 14 + nc
                        edges.append([node, neigh])

    edge_index = torch.tensor(edges).t().contiguous()
    return edge_index

EDGE_INDEX = create_grid_edges()
```

```python
from torch_geometric.data import Data

def image_to_graph(example):

    image = example["image"].convert("RGB")

    inputs = processor(image, return_tensors="pt")

    # FORCE GPU
    inputs = {k: v.to(device) for k, v in inputs.items()}

    # sanity check (REMOVE after debugging)
    assert inputs["pixel_values"].is_cuda, "Inputs are NOT on GPU!"

    with torch.no_grad():
        outputs = backbone(**inputs)

    patch_tokens = outputs.last_hidden_state[:, 1:, :].squeeze(0)

    return Data(
        x=patch_tokens.cpu(),
        edge_index=EDGE_INDEX,
        y=torch.tensor(example["label"])
    )
```

```python
import os
import torch

cache_dir = "/content/drive/MyDrive/endocellgraph/cache"
os.makedirs(cache_dir, exist_ok=True)

train_path = f"{cache_dir}/train_graphs.pt"
val_path = f"{cache_dir}/val_graphs.pt"
```

```python
if os.path.exists(train_path):
    train_graphs = torch.load(train_path)
    print("Loaded cached train graphs")
else:
    train_graphs = [image_to_graph(x) for x in train_ds]
    torch.save(train_graphs, train_path)
```

```python
if os.path.exists(val_path):
    val_graphs = torch.load(val_path)
    print("Loaded cached val graphs")
else:
    val_graphs = [image_to_graph(x) for x in val_ds]
    torch.save(val_graphs, val_path)
```
